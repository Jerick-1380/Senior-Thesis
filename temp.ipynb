```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_two_heatmaps(data1, data2, title1, title2, colormap='viridis'):
    # Create a figure with 1 row and 2 columns for subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot the first heatmap
    sns.heatmap(data1, cmap=colormap, cbar=True, ax=ax1)
    ax1.set_title(title1)

    # Plot the second heatmap
    sns.heatmap(data2, cmap=colormap, cbar=True, ax=ax2)
    ax2.set_title(title2)

    # Adjust layout for clean display
    plt.tight_layout()
    
    # Show the plot
    plt.show()

#[1,2,3,4,5]
data1 = [
    [91, 4, 2, 3, 0],
    [12, 87, 0, 1, 0],
    [0, 0, 100, 0, 0],
    [1, 7, 0, 84, 8],
    [2, 0, 0, 1, 97]
]
data2 = [
    [99, 1, 0, 0, 0],
    [37, 63, 0, 0, 0],
    [0, 0, 100, 0, 0],
    [0, 0, 0, 61, 39],
    [0, 0, 0, 3, 97]
]

plot_two_heatmaps(data1, data2, "Llama 3", "GPT-4o")
```
